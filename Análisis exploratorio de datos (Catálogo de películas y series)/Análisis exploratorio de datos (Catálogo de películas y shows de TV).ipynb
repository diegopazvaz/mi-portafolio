## 3. Limpieza de datos: Manejo de los datos faltantes

Hay varias columnas con valores vacíos. Sin embargo, eliminar las filas donde se encuentran esos valores no es una buena opción, ya que sería posible perder otros datos valiosos que se encuentren en esas mismas filas, es por eso que llenaré los datos vacíos dependiendo de la conclusión a la que quiero llegar con cada columna.

En la columna "duration" voy a identificar si los valores faltantes son de películas, de shows o de ambas.

```python
duracion_vacia = df[df["duration"].isnull()]
duracion_vacia
```

Todos los valores vacíos pertencen a registros de películas pero por alguna razón la duración de esas películas está registrada en la columna "rating", por lo que voy a crear una copia del dataframe y en él moveré estos datos a la columna a la que pertencen.

```python
df_completo = df
df_completo.loc[(df["duration"].isna()), "duration"] = duracion_vacia["rating"]
df_completo
```

La columna "duration" ahora está completa y tiene 0 valores faltantes.

```python
df_completo.isnull().sum()
```

Ahora voy a vaciar los valores de duración que estaban incorrectamente registrados en la columna "rating".

```python
df_completo.loc[duracion_vacia.index, 'rating'] = np.nan
df_completo.loc[duracion_vacia.index]
```

Una vez hecho esto, hay 7 valores faltantes en la columna "rating".

```python
df_completo.isnull().sum()
```

Con respecto a las columnas "director", "cast", "country", "date_added" and "rating", llenaré los valores vacíos con la palabra "Unknown" que significa "desconocido" en inglés, ya que todos los datos del dataframe están en ese idioma. Decidí llenarlos de esta manera debido a que más adelante quiero obtener los valores más repetidos de cada columna y de esta forma los valores faltantes no afectarán a la cuenta.

```python
nuevos_valores = {"director":"Unknown", "cast":"Unknown", "country":"Unknown", "date_added":"Unknown", "rating":"Unknown"}
df_completo = df.fillna(nuevos_valores)
```

```python
df_completo
```

Ahora ya no hay valores faltantes en el dataframe.

```python
df_completo.info()
```

```python
df_completo.isnull().sum()
```

## 5. Porcentaje de películas y shows de televisión en el catálogo

Lo primero que quiero saber es cómo están distribuidas las películas y shows de televisión en el catálogo, es por eso que contaré cuántas de cada una hay.

Hay 6131 películas y 2676 shows de TV. Ahora mostraré estos resultados en una gráfica de pie, ya que ayuda a ver la distribución del catálogo completo y el porcentaje que cada tipo de contenido representa. También es muy útil para comparar fácilmente ambas categorías.

La mayoría del catálogo está conformado por películas, las cuales representan el 69.6% de él, mientras que el 30.4% restante corresponde a los shows de TV.

## 6. Director con más contenido en el catálogo

Para saber quién es el director con más contenido publicado en el catálogo voy a obtener el número de producciones que tiene cada director que aparece en él y los ordenaré descendentemente a partir de ese número. Sin embargo, es posible que algunos contenidos hayan sido dirigidos por más de un director, por lo que me aseguraré de que todos los directores aparezcan registrados individualmente.

```python
df_directores = df_catalogo.assign(director=df_catalogo['director'].str.split(', ')).explode('director')
df_directores.info()
```

En el dataframe llamado "df_directores" tengo 9612 registros, ya que todos los directores están registrados individualmente, es decir que si una película tiene dos directores habrán dos registros de esa película, en uno de ellos aparecerá uno de los directores registrados y en el segundo registro aparecerá el otro director registrado.

```python
agrupar_directores = df_directores.groupby(["director"])["director"].count().sort_values(ascending=False)
agrupar_directores.head(11)
```

El director con más producciones en el catálogo es Rajiv Chilaka pero es importante señalar que hay 2634 producciones que tienen un director desconocido, es importante investigar la razón por la qué estos campos no fueron llenados.

Para mejorar la visualización crearé un gráfico de barras con el top 10 de los directores con más contenido en el catálogo, sin tomar en cuenta los directores desconocidos. Este tipo de gráfico facilita la comparación, puede ser ordenado de mayor a menor y permite ver tendencias. 

```python
fig, ax = plt.subplots(figsize=(6,5))
barras_directores = ax.bar(agrupar_directores[1:11].index, agrupar_directores[1:11].values, color='#afc0d8')

for barra in barras_directores:
    ax.text(barra.get_x() + barra.get_width()/2,
            barra.get_height() + 0.5,
            str(int(barra.get_height())),
            ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.title('Top 10 directores con más contenido en el catálogo', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Director', labelpad=20, fontsize=12, fontweight='bold')
plt.ylabel('Número de producciones', labelpad=20, fontsize=12, fontweight='bold')
plt.yticks(np.arange(2, 25, 2))
ax.set_xticks(agrupar_directores[1:11].index)
ax.set_xticklabels(agrupar_directores[1:11].index, rotation=90)
ax.grid(axis='y', linestyle='-', alpha=0.3)
ax.set_axisbelow(True);
```

Cada barra representa un director y la altura de la barra es el número de producciones que tiene. El gráfico muestra con claridad que la diferencia de contenido entre los 10 directores es poca, incluso hay dos empates en el cuatro lugar y entre los últimos tres directores.

## 7. Distribución de la clasificación de las películas y de los shows de TV

Cada producción agregada en el catálogo está clasificada de acuerdo al público al que va dirigida. Por lo que comenzaré haciendo una lista de todas las clasificaciones que están presentes en el catálogo y contando el número de películas y shows que tiene cada una.

```python
agrupar_rating = df_catalogo.groupby(["rating"])["rating"].count().sort_values(ascending=False)
agrupar_rating
```

También quiero saber el porcentaje que cada una de las clasificaciones representa en el catálogo, así que voy a calcularlo.

```python
porcentaje = (agrupar_rating / agrupar_rating.sum() * 100).round(2)
porcentaje
```

Para mejorar la visualización voy a concatenar las dos series que obtuvé previamente y crearé un nuevo dataframe.

```python
df_porcentaje_rating = pd.concat([agrupar_rating, porcentaje], axis=1)
df_porcentaje_rating.columns = ['Total', 'Porcentaje']
df_porcentaje_rating = df_porcentaje_rating.reset_index().rename(columns={'rating': 'Clasificación'})
df_porcentaje_rating
```

En el catálogo hay 12 clasificaciones diferentes, 7 producciones tienen una clasificación desconocida y es importante buscar la razón por la que estos datos no fueron registrados, también hay 83 producciones que no tienen clasificación (UR o NR) lo cual no es bueno pues no se sabe hacia que público están dirigidas. Lo que más me llamo la atención de este dataframe es que hay dos clasificaciones que en conjunto representan el 60.94% del catálogo, las cuales son TV-MA y TV-14.

## 8. Relación entre el top 10 de directores con más producciones y la clasificación de estas

Ahora que he identificado quienes son los 10 directores con más producciones en el catálogo y la distribución de las clasificaciones dentro de este, quiero saber que clasificación tienen las películas y shows de TV que han hecho estos directores. Para hacer esto crearé un dataframe que solo contiene las producciones de estos directores.

Con el objetivo de ver claramente la relación entre las clasificaciones y los directores voy a mostrar los datos en un gráfico de barras apiladas, esto facilita la visualización de variables númericas (número de producciones) con variables categóricas (director y clasificación). El único director que tiene cambios significativos en las clasificaciones de sus producciones es Martin Scorsese, quien ha hecho que algunas sean aptas para el público general con supervisión de los padres (PG), mientras que otras están dirigidas específicamente a adultos y bajo ninguna circunstancia cualquier persona menor de 18 años debe verlas (TV-MA y R).

## 10. Películas y shows de TV agregados cada año

Es importante saber si cada año el número de producciones agregadas al catálogo va aumentando o disminuyendo.

Primero voy a revisar si los registros de la columna "date_added" son strings o si tienen algún formato de fecha.

```python
df_catalogo.info()
```

Los registros son strings (object), por lo que los debo convertir a un formato de fecha. Para poder hacer esto primero tengo que asegurarme de que no tengan espacios al inicio y al final. También cambiaré todos los valores "Unknown" a "NaT" que es el equivalente de null cuando se trabaja con fechas.

```python
df_catalogo["date_added"] = df_catalogo["date_added"].replace("Unknown", None)
df_catalogo["date_added"] = df_catalogo["date_added"].str.strip()
df_catalogo.info()
```

Una vez hechos los dos pasos anteriores, realizaré el cambio de estos strings a un formato de fecha.

```python
df_catalogo['date_added'] = pd.to_datetime(df_catalogo['date_added'], format="%B %d, %Y")
df_catalogo
```

Es importante resaltar que todos los valores que solían ser "Unknown" ahora son "NaT", que significa lo mismo que null pero que hace posible hacer la conversión de los demás valores sin provocar un error.

```python
df_catalogo.loc[df_catalogo['date_added'].isnull()]
```

Ahora es posible extraer el año de cada una de las fechas y crear una serie de pandas con ellos.

```python
años_agregadas = df_catalogo['date_added'].dt.year
años_agregadas
```

Utilizando esta serie contaré el número de películas y shows que fueron agregadas año tras año.

```python
conteo_años = años_agregadas.value_counts().sort_index(ascending=True)
conteo_años
```

Mi objetivo es ver si, año tras año, el número de producciones añadidas tiende a aumentar o disminuir. Un gráfico de líneas es ideal para mostrar datos de tendencias y rastrear cambios a lo largo del tiempo. Por eso, crearé uno para mostrar los datos obtenidos.

```python
fig, ax = plt.subplots(figsize=(6,5))
ax.plot(conteo_años.index, conteo_años.values, marker='o', linestyle='-', color='#236424')
ax.set_xlabel("Año", fontweight='bold', labelpad=20, fontsize=12)
ax.set_ylabel("Producciones agregadas", fontweight='bold', labelpad=20, fontsize=12)
ax.set_title("Producciones agregadas por año", fontsize=18, fontweight='bold', pad=20)
ax.grid(axis='both', linestyle='-', alpha=0.3)
ax.set_axisbelow(True)
ax.set_ylim([0, 2500])
ax.set_xticks(conteo_años.index)
ax.tick_params(axis="x", rotation=90)
for i, value in enumerate(conteo_años.values):
    if value < 400:
        ax.text(conteo_años.index[i], value + 60, str(value), ha='center', va='bottom', fontsize=8);
    elif value == 1498:
        ax.text(conteo_años.index[i], value - 75, str(value), ha='right', va='center', fontsize=8);
    elif value == 1879:
        ax.text(conteo_años.index[i], value + 65, str(value), ha='left', va='center', fontsize=8);
    else:
        ax.text(conteo_años.index[i], value + 60, str(value), ha='right', va='center', fontsize=8);
```

En este gráfico, tengo todos los años ordenados ascendentemente en el eje x, mientras que en el eje y tengo el número de producciones añadidas. Cada punto representa el número de películas o shows de TV añadidos en cada año. De 2008 a 2012 hubo muchos cambios en el número de producciones añadidas por año; el más significativo ocurrió en 2011 y 2012, con una disminución del 76.92%. Sin embargo, de 2013 a 2019 la situación fue completamente diferente, ya que año tras año hubo un aumento en el número de producciones añadidas. Lo que me llamó la atención es que en los dos últimos años registrados en el dataframe (2020 y 2021) este número ha disminuido, lo cual no es necesariamente algo malo, pero es importante profundizar en ello para comprender qué está sucediendo.

## 14. Distribución de los géneros cinematográficos en el catálogo

También quiero saber como están distribuidos los géneros cinematográficos en el catálogo. Algunas películas y shows pertenecen a más de un género, es por eso que primero separaré cada uno de ellos en un registro diferente.

```python
df_generos = df_catalogo.assign(listed_in=df_catalogo['listed_in'].str.split(', ')).explode('listed_in')
df_generos.head(10)
```

Ahora hay un registro de cada producción para cada género que tiene asignado, gracias a esto puedo contar cuantas veces aparece cada género en el catálogo.
